Pandas - Series

```python
# Pandas 라이브러리 로딩
import pandas as pd

# 설치된 버전 출력
print(pd.__version__)

'''
# 리스트 데이터로 Series 객체 생성하기
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
year = ['2019', '2020', '2021', '2022']
result = pd.Series(data=year)

print('\nType:', type(result))
print(result)


'''
# Series 객체의 주요 속성
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
print('\nIndex: ', result.index )
print('Data: ', result.values )
print('DType: ', result.dtype )
print('Shape: ', result.shape )
print('\n')


'''
# Series 객체의 이름 지정
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
result.name = 'Year'
result.index.name = 'No.'
print(result)
print('\n')


'''
# 인덱스 추가하여 Series 객체 생성하기
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
year = ['2019', '2020', '2021', '2022']
idx = ['a', 'b', 'c', 'd']
result = pd.Series(data=year, index=idx, name='Year')
print(result)
print('\n')


'''
# 딕셔너리 데이터로 Series 객체 생성하기
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
score = {'Kim':85, 'Han':89, 'Lee':99, 'Choi':70}
result = pd.Series(data=score, name='Score')
print(result)
```

DataFrame 데이터 살펴보기

```python
import pandas as pd

# 실습 데이터 생성
score = {'name': ['Jessi', 'Emma', 'Alex', 'Jessi', 'Tom'],
         'score': [100, 95, 80, 85, 97],
         'grade': ['A', 'A', 'B', 'B', 'A'],
         'subject':['python', 'java', 'python', 'c', 'java']}

score_df = pd.DataFrame(data=score)
print(score_df)


'''
.info() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n<Summary of DataFrame>')
print(score_df.info())


'''
.head() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n<Return the first n rows>')
print(score_df.head(3))


'''
.tail() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n<Return the last n rows>')
print(score_df.tail(2))


'''
.sample() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n<Return a random sample>')
print(score_df.sample())


'''
.sample() 함수의 random_state 인자 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n<Return a random sample with random_state>')
print(score_df.sample(2, random_state=10))


'''
.sample() 함수의 flac 인자 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n<Return a random sample with flac 50%>')
print(score_df.sample(frac=0.5))
```

수치형/범주형 데이터

```python
import pandas as pd

# 실습 데이터 생성
score = {'name': ['Jessi', 'Emma', 'Alex', 'Jessi', 'Tom'],
         'age': [20, 24, 23, 20, 27],
         'score': [100, 95, 80, 85, 97],
         'grade': ['A', 'A', 'B', 'B', 'A'],
         'subject':['python', 'java', 'python', 'c', 'java']}

score_df = pd.DataFrame(data=score)
print(score_df)


'''
describe() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print("\nGenerate Descriptive Statistics")
print(score_df.describe())


'''
범주형 데이터에 대한 describe() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print("\nGenerate Descriptive Statistics for Categorical")
print(score_df[['grade','subject']].describe())


'''
describe(include='all') 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print("\nGenerate Descriptive Statistics")
print(score_df.describe(include='all'))


'''
범주형 데이터를 위한 unique() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print("\nThe Unique Values")
print(score_df['subject'].unique())


'''
범주형 데이터를 위한 value_counts() 함수 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print("\nThe Value Counts")
print(score_df['subject'].value_counts())


print("\nThe Value Counts-normalize")
print(score_df['subject'].value_counts(normalize=True))
```

```python
import pandas as pd

# 실습 데이터 생성
score = {'name': ['Jessi', 'Emma', 'Alex', 'Jessi', 'Tom'],
         'score': [100, 95, 80, 85, 97],
         'grade': ['A', 'A', 'B', 'B', 'A'],
         'subject':['python', 'java', 'python', 'c', 'java']}
c = ['name', 'subject', 'score', 'grade', 'etc']

df = pd.DataFrame(data=score, columns=c)
print(df)


'''
Series 객체 생성하여 'semester' 라는 열 추가
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
semester_data = pd.Series(['20-01', '20-01', '20-02', '20-01'])
df['semester'] = semester_data

print('\nAdd Columns : semester')
print(df)


'''
연산 이용하여 'high_score' 라는 열 추가
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
df['high_score'] = df['score'] > 90 

print('\nAdd Columns : high_score')
print(df)


'''
6 번 인덱스에 행 데이터 추가
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
df.loc[6] = ['Jina', 'python', 100, 'A', 1, '20-02', True]

print('\nAdd Row 6')
print(df)


'''
1) drop() 이용하여 행 데이터 삭제
2) inplace 인자를 사용하여 원본 데이터에 바로 삭제
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
df.drop(6, inplace=True) 

print('\nDelete Row : 6')
print(df)


'''
drop() 이용하여 'etc' 열 삭제 (원본데이터변경)
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
df.drop(columns=['etc'], inplace=True)

print('\nDelete Column :')
print(df)
```

결측치 처리 (2)

```python
import pandas as pd
import numpy as np

# 실습 데이터 생성
df = pd.DataFrame(data = np.arange(18).reshape(6,3),
				  index = ['a','b','c','d','e','f'],
                  columns=['col1','col2','col3'])
                  
df['col4'] = pd.Series(data = [1.7, 1.2, 2.4], 
                       index = ['a','e','c'])
df.loc['c'] = None
print('Sample Data')
print(df)

 
'''
1) 행을 기준으로 결측치 조회하여 제거하기
2) how=all 인자로 전체 nan 인 경우만 제거하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#1 Remove missing values')
print(df.dropna(how='all'))


'''
원본 데이터에 반영하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#2 Remove missing values with inpalce=True')
df.dropna(how='all', inplace=True)
print(df)

 
'''
열을 기준으로 결측치 조회하여 제거하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#3 Remove missing values in columns')
print(df.dropna(axis='columns'))

 
'''
주석(#)만 지우시고 실습 데이터를 추가하세요.
'''
df.iloc[:2, 0] = np.nan
df.iloc[:4, 1] = np.nan
print('\nSample Data')
print(df)


'''
결측값을 0 으로 대체하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#4 Replace missing values')
print(df.fillna(0))
'''
** 0이 실수인 이유 - 해당 시리즈와 같은 데이터 타입이어야 함
'''

'''
딕셔너리를 활용한 컬럼별 대체값 지정하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#5 Replace missing values')
replace_set = {'col2':0,'col4':'100'}
print(df.fillna(replace_set))


'''
1)대표값인 평균을 활용하여 결측치 대체하기
2)원본 데이터에 반영하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#6 Replace missing values')
replace_set = {'col1':df['col1'].mean()}
df.fillna(replace_set, inplace=True)
print(df)
```

데이터 조회 : 인덱싱

```python
import pandas as pd

# 실습 데이터 생성
d = {'name': ['Jessi', 'Emma', 'Alex', 'Jessi', 'Tom'],
     'score': [100, 95, 80, 85, 97],
     'grade': ['A', 'A', 'B', 'B', 'A'],
     'subject':['python', 'java', 'python', 'c', 'java']}

df = pd.DataFrame(data=d)
print('Sample Data')
print(df)


'''
name 컬럼을 행 인덱스로 바꾸기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#1 Set Index')
df.set_index('name', inplace=True)
print(df)
  

'''
loc[] 를 통해 행 데이터 조회하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#2 Access rows')
subset = df.loc[['Jessi', 'Emma']]
print(subset)


'''
iloc[] 를 통해 행 데이터 조회하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#3 Access rows by index position')
subset = df.iloc[[0,4]]
print(subset)


'''
loc[] 의 활용 : Boolean Indexing
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#4 Access rows by boolean indexing')
subset = df.loc[df['score'] >= 95]
print(subset)


'''
loc[] 의 활용 : Boolean Indexing
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#5 Access rows by boolean indexing')
subset = df.loc[df['subject'] == 'python']
print(subset)
```

데이터 슬라이싱

```python
import pandas as pd

# 실습 데이터 생성
d = {'name': ['Jessi', 'Emma', 'Alex', 'Jessi', 'Tom'],
     'score': [100, 95, 80, 85, 97],
     'grade': ['A', 'A', 'B', 'B', 'A'],
     'subject':['python', 'java', 'python', 'c', 'java']}

sample_df = pd.DataFrame(data=d)
sample_df.set_index('name', inplace=True)

print('Sample Data')
print(sample_df)


'''
기본 슬라이싱 기법을 통한 행 데이터 조회#1
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#1 Access Rows by slicing')
subset = sample_df[1:4]
print(subset)


'''
기본 슬라이싱 기법을 통한 행 데이터 조회#2
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#2 Access Rows by slicing')
subset = sample_df[::2]
print(subset)


'''
기본 슬라이싱 기법을 통한 행 데이터 조회#3
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#3 Access Rows by slicing')
subset = sample_df[:'Alex']
print(subset)


'''
loc 프로퍼티 활용한 슬라이싱 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#4 Access Rows by loc[slicing]')
subset = sample_df.loc[:'Emma', ['subject' , 'grade']] 
print(subset)


'''
iloc 프로퍼티 활용한 슬라이싱 실습
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#5 Access Rows by iloc[slicing]')
subset = sample_df.iloc[:4 , -1]
print(subset)
```

Pandas의 집계함수

```python
import pandas as pd
import numpy as np

# 실습 데이터 생성
score = {'sub1': [3, 9, 1, 1, 9],
         'sub2': [2, 9, np.nan, np.nan, 8],
         'sub3': [np.nan, 1, 5, 5, 7],
         'sub4': [np.nan, 3, np.nan, 1, np.nan]}

df = pd.DataFrame(data=score)
print(df)


'''
count() 집계함수 실습
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
print('\n#1 Count Function')
print(df.count())


'''
count 함수에 axis=1 인지 추가한 실습
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
print('\n#2 Count Function with axis=1')
print(df.count(axis=1))


'''
sum() 집계함수 실습
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
print('\n#3 Sum Function')
print(df.sum())


'''
sum() 함수에 skipna=False 인자 추가한 실습
(아래 주석#을 지우고 _____ 란에 실습을 시작하세요.)
'''
print('\n#4 Sum Function with skipna')
print(df.sum(skipna=False))
```

데이터의 그룹화

```python
import pandas as pd
import numpy as np

# 실습 데이터 생성
sample = {'product':['a','b','a','b','a','b','a','a'],
          'sensor':['s1','s1','s2','s3','s2','s2','s1','s3'],
          'x':np.arange(1,9),
          'y':np.arange(5,13)}

df = pd.DataFrame(data=sample)
print(df)


'''
product 컬럼 기준으로 그룹화하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#1 groupby-product')
grouped_product = df.groupby('product')
print(grouped_product)


'''
grouped_product 에 저장된 key와 value 출력하기
주석(#)만 제거하고 실행해보세요.
'''
print('\n#2 key & value')
for key, value in grouped_product:
      print("-----------------------")
      print("key :", key)
      print("value :\n", value)


'''
groupby 결과에 집계함수 적용하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#3 groupby-product')
grouped_product = df.groupby('product').sum()
print(grouped_product)


'''
product, sensor 기준으로 그룹화하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#4 groupby-product,sensor')
grouped_product = df.groupby(['product', 'sensor']).sum()
print(grouped_product)


'''
그룹화한 뒤 x 컬럼에 대해서만 집계함수 적용하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#5 groupby-product,sensor')
grouped_product = df.groupby(['product', 'sensor'])['x'].sum()
print(grouped_product)


'''
그룹화한 뒤 컬럼별 다른 집계함수 적용하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#6 groupby-product,sensor')
condition = {'x':'max', 'y':'min'}
grouped_product = df.groupby(['product', 'sensor']).agg(condition)
print(grouped_product)
```

데이터 합치기

```python
import pandas as pd

# 실습 데이터 생성
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])                   

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])

df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])
                    
print('DF1')
print(df1)
print('\nDF2')
print(df2)
print('\nDF3')
print(df3)
print('\nDF4')
print(df4)


'''
df2, df1 합치기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#1 Concat df2, df1')
result = pd.concat([df2, df1, df3], ignore_index=True)
print(result)


'''
shape 이 서로 다른 df1, df4 합치기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#2 Concat df1, df4')
result = pd.concat([df1, df4])
print(result)


'''
슬라이싱을 통해 일부 데이터 합치기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#3 Concat df1, df4 by slicing')
result = pd.concat([df1[:2], df4[3:]])
print(result)

 
'''
행 기준으로 df1, df4 합치기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#4 Concat df1, df4 by axis=1')
result = pd.concat([df1, df4], axis=1, ignore_index=True)
print(result)

 
'''
교집합으로 데이터 결합하기
주석(#)을 지우고, 주어진 빈칸의 코드를 완성하세요.
'''
print('\n#5 Concat df1, df4 by join=inner')
result = pd.concat([df1, df4], join='inner')
print(result)
```
